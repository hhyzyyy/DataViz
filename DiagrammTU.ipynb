```python
modebar = dict({
    'scrollZoom': True,
    'displaylogo': False, 
    'modeBarButtonsToRemove': ['lasso2d', 'select2d'],
    'toImageButtonOptions': {'format': 'png', # one of png, svg, jpeg, webp
    'filename': 'Zwischengespeicherter_Plot',}
})

iwfColors =  ['black', 'white', 'rgb(159, 182, 196)', 'rgb(125, 102, 102)', 'rgb(153, 0, 0)']
iwfColors_ohneWeiß =  ['black', 'rgb(159, 182, 196)', 'rgb(125, 102, 102)', 'rgb(153, 0, 0)']

# Continues color Scales IWF
IWFgreyBlueToBlack = ['#dfe7ec', '#9fb6c4', '#000000']
IWF_Red_Fade = ['#990000','#ffe6e6']
IWF_GreyBlue_fade = ['#3a515f', '#9fb6c4', '#dfe7ec']
IWF_Brown_fade = ['#382e2e', '#7d6666', '#e8e3e3']
IWF_Black_fade = ['#000000', '#f2f2f2']

# Hab es mithilfe von hsl format normiert (rot ist am dunkelsten). https://www.w3schools.com/colors/colors_picker.asp
IWF_Brown_equalIntensityFade = ['#f4f1f1','#544545']
IWF_Red_equalIntensityFade = ['#ffe6e6','#990000']
IWF_Blue_equalIntensityFade = ['#eff3f5', '#3a515f']

#Symbols
iwfSymbols = ["circle", "square", "diamond", "triangle-up", "triangle-down", "cross", "x"]
FraunhoferFont = dict(family="Arial",size=11,color="black")
```

```python
df = px.data.gapminder().query("continent=='Americas'")
ls_first4 = df.country.unique()[:5]
df = df.loc[df.country.isin(ls_first4)]

def scatterPlot_line_IWF(df=df, x=df.columns[0], y = df.columns[1], groupby = None, 
                         x_tick0 = None, x_dtick = None, x_range = None, x_nticks = None, x_rangemode = 'normal',
                        y_tick0 = None, y_dtick = None, y_range = None, y_nticks = None, y_rangemode = 'normal'):
    fig = px.scatter(df, 
               x=x, 
               y=y,
               symbol = groupby,
               symbol_sequence =  iwfSymbols,
               color=groupby,
              color_discrete_sequence= iwfColors_ohneWeiß,
               hover_name=groupby,
                 hover_data={groupby: False}
              )
    
    fig.update_layout(font=FraunhoferFont,
#                  autosize=False,
#                     width=700,
#                     height=700 * 2/3,
                  plot_bgcolor ='white')
    fig.update_traces(mode='lines+markers',
                  line_shape='spline',
#                       marker_color = iwfColors_ohneWeiß,
                  marker_size = 6,
                  marker_line=dict(
                        color='black',
                        width=1)
                 )
    fig.update_xaxes(showline=True, linewidth=1.5,
                 linecolor='black', gridcolor='black', gridwidth=1,
                 zeroline=True, zerolinewidth=1.5, zerolinecolor='black',
                     tick0 = x_tick0,
                     dtick = x_dtick,
                     range = x_range,
                     nticks = x_nticks,
                     rangemode = x_rangemode
                    )
    fig.update_yaxes(showline=True, linewidth=1.5,
                 linecolor='black', gridcolor='black', gridwidth=1,
                 zeroline=True, zerolinewidth=1.5, zerolinecolor='black',
                     tick0 = y_tick0,
                     dtick = y_dtick,
                     range = y_range,
                     nticks = y_nticks,
                     rangemode = y_rangemode,
                    )
    
    
    fig.add_shape(
        # Rectangle with reference to the plot
            type="rect",
            xref="paper",
            yref="paper",
            x0=0,
            y0=0,
            x1=1.0,
            y1=1.0,
            line=dict(
                color="black",
                width=1,
            )
        )
    
    return fig

def scatterPlot_IWF(df=df, x=df.columns[0], y = df.columns[1], groupby = None):
    fig = px.scatter(data_frame = df, 
               x=x, 
               y=y,
               symbol = groupby,
               symbol_sequence =  iwfSymbols,
               color=groupby,
              color_discrete_sequence= iwfColors_ohneWeiß,
               hover_name=groupby,
                 hover_data={groupby: False}
              )
    
    fig.update_layout(font=FraunhoferFont,
#                  autosize=False,
#                     width=700,
#                     height=700 * 2/3,
                  plot_bgcolor ='white')
    fig.update_traces(
                  line_shape='spline',
#                       marker_color = iwfColors_ohneWeiß,
                  marker_size = 6,
                  marker_line=dict(
                        color='black',
                        width=1)
                 )
    fig.update_xaxes(showline=True, linewidth=1.5,
                 linecolor='black', gridcolor='black', gridwidth=1,
                 zeroline=True, zerolinewidth=1.5, zerolinecolor='black',          
                    )

    fig.update_yaxes(showline=True, linewidth=1.5,
                 linecolor='black', gridcolor='black', gridwidth=1,
                 zeroline=True, zerolinewidth=1.5, zerolinecolor='black'
                    )
    
    
    fig.add_shape(
        # Rectangle with reference to the plot
            type="rect",
            xref="paper",
            yref="paper",
            x0=0,
            y0=0,
            x1=1.0,
            y1=1.0,
            line=dict(
                color="black",
                width=1,
            )
        )
    
    return fig
    


fig = scatterPlot_line_IWF(df = df, x = 'year', y = 'lifeExp', 
                           groupby = 'country', 
                           x_range = [1950, 2010],
                            y_range = [40,90],
#                            y_tick0 = 0,
#                            y_dtick = 30                           
#                            y_rangemode = 'tozero'
                          )
fig.show()
```

```python
df = px.data.tips()

def boxPlot_IWF(df = None, x = None, y = None, groupby = None):

    fig = px.box(data_frame = df, x = x, y = y, color = groupby,
                color_discrete_sequence= iwfColors_ohneWeiß)
    fig.update_layout(font=FraunhoferFont,
#                  autosize=False,
#                     width=700,
#                     height=700 * 2/3,
              plot_bgcolor ='white') 
    
    fig.update_traces(marker_line=dict(
                        color='black',
                        width=1))
    fig.update_xaxes(showline=True, linewidth=1.5, linecolor='black', 
                     gridcolor='black', gridwidth=1,
         zeroline=True, zerolinewidth=1.5, zerolinecolor='black',
                    tickson = "boundaries")
    
    fig.update_yaxes(showline=True, linewidth=1.5,
             linecolor='black', gridcolor='black', gridwidth=1,
             zeroline=True, zerolinewidth=1.5, zerolinecolor='black')
    
    fig.add_shape(
        # Rectangle with reference to the plot
            type="rect",
            xref="paper",
            yref="paper",
            x0=0,
            y0=0,
            x1=1.0,
            y1=1.0,
            line=dict(
                color="black",
                width=1,
            ))
    return fig

fig = boxPlot_IWF(
                df = df,
                 x = 'day',
                 y = 'total_bill',
                 groupby = 'smoker'
    
    
)
    

# fig = px.box(df, x="day", y="total_bill", color="smoker")
# fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default
fig.show(config =  modebar)
```

```python
df = px.data.tips()

def violinPlot_IWF(df = None, x = None, y = None, groupby = None):

    fig = px.violin(data_frame = df, x = x, y = y, color = groupby,
                color_discrete_sequence= iwfColors_ohneWeiß)
    fig.update_layout(font=FraunhoferFont,
#                  autosize=False,
#                     width=700,
#                     height=700 * 2/3,
              plot_bgcolor ='white') 
    
    fig.update_traces(marker_line=dict(
                        color='black',
                        width=1))
    fig.update_xaxes(showline=True, linewidth=1.5, linecolor='black', 
                     gridcolor='black', gridwidth=1,
         zeroline=True, zerolinewidth=1.5, zerolinecolor='black',
                    tickson = "boundaries")
    
    fig.update_yaxes(showline=True, linewidth=1.5,
             linecolor='black', gridcolor='black', gridwidth=1,
             zeroline=True, zerolinewidth=1.5, zerolinecolor='black')
    
    fig.add_shape(
        # Rectangle with reference to the plot
            type="rect",
            xref="paper",
            yref="paper",
            x0=0,
            y0=0,
            x1=1.0,
            y1=1.0,
            line=dict(
                color="black",
                width=1,
            ))
    return fig

fig = violinPlot_IWF(
                df = df,
                 x = 'day',
                 y = 'total_bill',
                 groupby = 'smoker'
    
    
)
    

# fig = px.box(df, x="day", y="total_bill", color="smoker")
# fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default
fig.show(config =  modebar)
```
